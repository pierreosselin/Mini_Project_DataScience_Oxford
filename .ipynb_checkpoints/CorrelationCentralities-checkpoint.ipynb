```python
import numpy as np
from math import *
import networkx as nx
import matplotlib.pyplot as plt
from igraph import *
import pandas as pd
import seaborn as sns
from scipy.stats import linregress
from matplotlib.patches import Rectangle
```

```python
path = "/home/pierre/Documents/CourseOxford/DataScience/Problem_sheet_2/data_revised/"
lData = []
def importMat(i):
    mat = pd.read_csv(path + "adjmat" + str(i) + ".csv", header=None)
    return mat

for i in range(43):
    lData.append(importMat(i+1).values)

dfHouse = pd.read_csv("/home/pierre/Documents/CourseOxford/DataScience/Problem_sheet_2/data_revised/household_data.csv", header = None)
matHouse = dfHouse.values
lHouseVill = []
c = 0
for i in range(43):
    a = lData[i].shape[0]
    lHouseVill.append(matHouse[c:c + a])
    c += a
```

```python
countVillage = [0 for i in range(43)]
for household in matHouse:
    countVillage[household[0] - 1] += household[3]
for i in range(43):
    countVillage[i] /= lData[i].shape[0]
pd.DataFrame(countVillage)
"""
ax = plt.subplot(111, frame_on=False) # no visible frame
ax.xaxis.set_visible(False)  # hide the x axis
ax.yaxis.set_visible(False)  # hide the y axis

pd.tools.plotting.table(ax, pd.DataFrame(countVillage))  # where df is your data frame
"""
sns.distplot(countVillage, hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.savefig("DensityPlot")
```

```python
lGraph = []
for i in range(43):
    g = Graph.Adjacency(lData[i].tolist())
    lGraph.append(g)
```

```python
def mean(l):
    return sum(l)/(float(len(l)))

lCentrality = []
for i in range(43):
    a = lData[i].shape[0]
    index = [j for j in range(a) if lHouseVill[i][j,2] == 1]
    lm = lGraph[i].evcent(directed = False, scale = False)
    levCentralities = mean([lm[i] for i in index])
    lInDegree = mean(lGraph[i].indegree(vertices = index))
    lOutDegree = mean(lGraph[i].outdegree(vertices = index))
    lCloseness = mean(lGraph[i].closeness(vertices = index))
    print(lGraph[i].closeness(vertices = index))
    lBetweeness = mean(lGraph[i].betweenness(vertices = index))
    lPageRank = mean(lGraph[i].personalized_pagerank(vertices = index))
    lCentrality.append([levCentralities,lInDegree,lOutDegree,lCloseness,lBetweeness,lPageRank])
lCentrality = np.array(lCentrality)
df = pd.DataFrame(data = lCentrality, columns = ["ev", "InDegree", "OutDegree", "Closeness", "Betweeness", "PageRank"])
plt.figure(figsize=(20,10))
plt.subplot(3,2,1)
sns.distplot(lCentrality[:,0], hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'}, axlabel = "Mean Eigenvalue Centrality for the leaders")
plt.subplot(3,2,2)
sns.distplot(lCentrality[:,1], hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'}, axlabel = "Mean InDegree Centrality for the leaders")
plt.subplot(3,2,3)
sns.distplot(lCentrality[:,2], hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'}, axlabel = "Mean OutDegree Centrality for the leaders")
plt.subplot(3,2,4)
sns.distplot(lCentrality[:,3], hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'}, axlabel = "Mean Closeness Centrality for the leaders")
plt.subplot(3,2,5)
sns.distplot(lCentrality[:,4], hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'}, axlabel = "Mean Betweeness Centrality for the leaders")
plt.subplot(3,2,6)
sns.distplot(lCentrality[:,5], hist=True, kde=False, 
             bins=20, color = 'blue',
             hist_kws={'edgecolor':'black'}, axlabel = "Mean PageRank Centrality for the leaders")
plt.savefig("CentralityPlots")
```

```python
plt.figure(figsize=(20,15))
lName = ["Eigenvector Centrality", "InDegree Centrality", "OutDegree Centrality", "Closeness Centrality", "Betweeness Centrality", "PageRank Centrality"]
for i in range(6):
    plt.subplot(3,2,i + 1)
    plt.xlabel(lName[i])
    plt.ylabel("Participation Ratio")
    gradient, intercept, r_value, p_value, std_err = linregress(lCentrality[:,i],countVillage)
    
    mn=np.min(lCentrality[:,i])
    mx=np.max(lCentrality[:,i])
    plt.text(mn,0.4,"R = " + "{:.3f}".format(r_value) + "      pValue = " + "{:.3f}".format(p_value), size= 14)
    x1=np.linspace(mn,mx,500)
    y1=gradient*x1+intercept
    plt.plot(lCentrality[:,i],countVillage, '.')
    plt.plot(x1,y1)
plt.savefig("Regressions")
```
