```python
def proportionAdoption(index, lHouseVill):
    prop = 0
    n = 0
    for i in range(lHouseVill[index].shape[0]):
        n += 1
        if lHouseVill[index][i,-1] == 1:
            prop += 1
    return (prop/n)
```

### Households adopted microfinance more likely have relationships together?

```python
## G = NetworkX graph, Partition = list of list of node
def Modularity(G, partition):
    indices = dict(zip(list(G.nodes()), [i for i in range(len(G))]))
    m = G.number_of_edges()
    K = dict(list(G.degree()))
    A = nx.adjacency_matrix(G)
    result = 0
    for com in partition:
        for node1 in com:
            for node2 in com:
                result += A[indices[node1], indices[node2]] - (K[node1]*K[node2])/(2*m)
    return (result/(2*m))
```

### Homophily

```python
## Mixing matrices
def MixingMatrix(G,lHouse):
    n = G.number_of_nodes()
    m = G.number_of_edges()
    n_interAdopt = 0
    n_interNonAdopt = 0
    n_intraAdoptNonAdopt = 0
    lEdges = G.edges()
    for edge in lEdges:
        if lHouse[edge[0], -1] != lHouse[edge[1], -1]:
            n_intraAdoptNonAdopt += 1
        elif lHouse[edge[0], -1] == 1:
            n_interAdopt += 2
        else :
            n_interNonAdopt += 2
    n_interAdopt, n_interNonAdopt, n_intraAdoptNonAdopt = n_interAdopt/(2*m), n_interNonAdopt/(2*m), n_intraAdoptNonAdopt/(2*m)
    interm = (n_interAdopt + n_intraAdoptNonAdopt)**(2) + (n_interNonAdopt + n_intraAdoptNonAdopt)**(2)
    homo = (n_interAdopt + n_interNonAdopt - interm)/(1-interm)
    return homo
```

```python
listHomo = []
for i in range(43):
    G = nx.from_numpy_matrix(lData[i])
    lol = lHouseVill[i]
    listHomo.append(MixingMatrix(G, lol))
lFinal = sorted(range(len(listHomo)), key = lambda k : listHomo[k])
lFinal.reverse()
print(lFinal)
print(listHomo[10])
plt.xlabel("Assortative Coefficients")
plt.ylabel("Cardinality")
plt.title("Density Plot for the Assortative Coefficients")
sns.distplot(listHomo, hist=True, kde=False, 
             bins=40, color = 'blue',
             hist_kws={'edgecolor':'black'})
mu, std = norm.fit(listHomo)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
print("Fit results: mu = %.2f,  std = %.2f" % (mu, std))
plt.savefig("DensityPlotHomophily", dpi=900)
plt.show()
```

### Compare the p inter adoption with the global p

```python
listResult = []
for i in range(43):
    mat = lData[i]
    G = nx.from_numpy_matrix(mat)
    m = G.number_of_edges()
    n = G.number_of_nodes()
    globalp = (2*m)/(n*(n-1))
    lol = lHouseVill[i]
    listIndex1 = []
    listIndex2 = []
    for j in range(lol.shape[0]):
        if lol[j, -1] == 1:
            listIndex1.append(j)
        else:
            listIndex2.append(j)
    subGraphAdoption = mat[np.ix_(listIndex1,listIndex1)]
    subG = nx.from_numpy_matrix(subGraphAdoption)
    m2 = subG.number_of_edges()
    n2 = subG.number_of_nodes()
    p2 = (2*m2)/(n2*(n2-1))
    print(i, p2/globalp)
    listResult.append((p2/globalp))
sns.distplot(listResult, hist=True, kde=False, 
             bins=40, color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.savefig("DensityPlot")
```

### Question 2 : Villages with communities that align with adoption

### Given partition gives insight into the structure

```python
def partitionAnal(dictPart, lHouse):
    lFraction = []
    lMax = []
    lValues = list(dictPart.values())
    dUse = {}
    for val in lValues:
        dUse[val] = []
    for node, val in dictPart.items():
        dUse[val].append(node)
    for part, lNodes in dUse.items():
        denom = 0
        n = len(lNodes)
        if n > 1:
            for node in lNodes:
                if lHouse[node, -1] == 1:
                    denom += 1
            lFraction.append((n,denom/n))
            lMax.append((denom/n, n))
    maxEl = max(lMax)
    return [lFraction, maxEl]
```

```python
def distributionPropCluster(lCluster):
    lProp = []
    for el in lCluster:
        lProp.append(el[1])
    print(lProp)
    sns.distplot(lProp, hist=True, kde=False, 
             bins=40, color = 'blue',
             hist_kws={'edgecolor':'black'})
```

```python
lResult = []
lMean = []
lMax = []
for i in range(43):
    mat = lData[i]
    G = nx.from_numpy_matrix(mat)
    dictCom = community.best_partition(G)
    [lFraction, maxEl] = partitionAnal(dictCom, lHouseVill[i])
    lResult.append(lFraction)
    maxEl = tuple((list(maxEl)) + [i])
    lMax.append(maxEl)
lMaxSorted = sorted(lMax, key=lambda k: k[0])
print(lResult)
for i in range(43):
    part = lResult[i]
    mean = 0
    for el in part:
        if el[1] > 0.5:
            mean += el[1]
        else:
            mean += 1 - el[1]
        
    mean /= (len(part))
    lMean.append(mean)
print(lMean)
listFinal = sorted(range(len(lMean)), key = lambda k : lMean[k])
listFinal.reverse()
listFinal
```
